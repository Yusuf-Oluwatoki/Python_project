```
S_i = linspace(start=100,stop=2000,num=20)
S_i
```

```
jam=[]
albino=[]
for i in S_i:
    
    albino.append(i)
    Te = (((1-alpha)*i)/(4*sigma*epsilon))**0.25 
    Temp=Te-273
    jam.append(Temp)
    
    #scatter(Temp,i)
    
    
```

```

import numpy as np
import matplotlib.pyplot as plt
import math


e = 0.61
S = 1372
b = 5.67*math.pow(10,-8)


s = 0.61
y = 0.3

'''QUESTION 2'''

T_int= 253
T_int2 = 263
T_int3 = 273
T_int4 = 283
T_int5 = 293

n=401
dt=10**21

cp1=1.05*10**23


T_new5=[]
T_new4=[]
T_new3=[]
T_new2=[]
T_new=[]

time=[]


for i in range(n):
    t=i*dt
    time.append(t)

for i in range(n):
    Temp=(S*(1-y)*dt-4*s*b*(T_int**4)*dt)/cp1+T_int
    T2=(S*(1-y)*dt-4*s*b*(T_int2**4)*dt)/cp1+T_int2
    T3=(S*(1-y)*dt-4*s*b*(T_int3**4)*dt)/cp1+T_int3
    T4=(S*(1-y)*dt-4*s*b*(T_int4**4)*dt)/cp1+T_int4
    T5=(S*(1-y)*dt-4*s*b*(T_int5**4)*dt)/cp1+T_int5
    
    T_new.append(Temp)
    T_new2.append(T2)
    T_new3.append(T3)
    T_new4.append(T4)
    T_new5.append(T5)
    
    T_int = Temp
    T_int2 = T2
    T_int3 = T3
    T_int4 = T4
    T_int5 = T5

plt.plot(time,T_new)
plt.plot(time[40],T_new[40],'rs', label = T_new[40])

plt.plot(time,T_new2)
plt.plot(time[40],T_new2[40],'bs', label = T_new2[40])

plt.plot(time,T_new3)
plt.plot(time[40],T_new3[40],'ys', label = T_new3[40])

plt.plot(time,T_new4)
plt.plot(time[40],T_new4[40],'gs', label = T_new4[40])

plt.plot(time,T_new5)
plt.plot(time[40],T_new5[40],'ms', label = T_new5[40])

plt.legend()
plt.grid()
figsize(15,5)
```
