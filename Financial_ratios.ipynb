```
import numpy as np
import pandas as pd
#!pip install xlrd
```

```
df = pd.read_csv('flour_mill1.csv',index_col=0).T
#df=df.drop(columns=['Unnamed: 1'], axis=1)
df=df.fillna(0)
#df.replace(u'\xa0', '')
df.head()
```

```
cols = df.columns
#cols.remove('fistcolumn')
for col in cols:
    df[col] = df[col].astype(float)
```

```
df.index
```

```
column_list=df.columns
column_list
```

```

Cost_to_sales=-df['Cost of sales']/df['Sales']*100
Gross_profit=df['Cost of sales']+df['Sales']
Gross_margin=Gross_profit/df['Sales']*100
RTS=-df[' Royalties, Technical Service & Mgt fees ']/df['Sales']*100
AaP=-df[' Advert,Promo & distibution ']/df['Sales']*100
Admin=-df['Admin cost']/df['Sales']*100
Op_profit= Gross_profit + df[' Advert,Promo & distibution '] + df['Admin cost']+df['Other income']
EBITDA=Op_profit-df['Depreciation/ Amortisation']  
EBITDA_margin=EBITDA/df['Sales'] *100
PPE=[np.mean(df['PP&E'][:1]),np.mean(df['PP&E'][0:2]),np.mean(df['PP&E'][1:3]),np.mean(df['PP&E'][2:4]),np.mean(df['PP&E'][3:5])]
Depreciation=-df['Depreciation/ Amortisation']/PPE*100
Net_interest=df['Interest Received']+df['Interest paid']
cash=[np.mean(df['Cash'][:1]),np.mean(df['Cash'][0:2]),np.mean(df['Cash'][1:3]),np.mean(df['Cash'][2:4]),np.mean(df['Cash'][3:5])]
Interest_recieved=df['Interest Received']/cash*100
Gross_debt=df['ST debt']+df['LT debt']
Gross_debt_t=[np.mean(Gross_debt[:1]),np.mean(Gross_debt[0:2]),np.mean(Gross_debt[1:3]),np.mean(Gross_debt[2:4]),np.mean(Gross_debt[3:5])]
Interest_paid=-df['Interest paid']/Gross_debt_t*100
EBT=Op_profit+Net_interest
EBT_after=EBT+df['Income from associate']+df[' Exceptional items/ Doubtful debt provisions, % of sales ']
Tax_rate=-df['Tax']/EBT_after*100
PAT=df['Tax']+EBT_after
Net_income=PAT+df['Extraordinary items']+df['Minority Interest1']+df['Preference Dividend']
Pay_out=-df['Ordinary Dividend']/Net_income*100
Retained=df['Ordinary Dividend']+Net_income
Adjusted=df['Basic']+df['Additional/Bonus']
EPS=Net_income/Adjusted
DPS=df['Ordinary Dividend']/Adjusted
Shareholder=df['Issued capital']+df['Share premium']+df['Reserves']+df['Retained earnings']
BVPS=Shareholder/Adjusted
CFPS=df['Net cash from operations']/Adjusted
```

```
dff=pd.DataFrame(Cost_to_sales, columns = ['Cost_to_sales'])
dff['Gross_profit']=Gross_profit
dff['Gross_margin']=Gross_margin
dff['RTS']=RTS
dff['AaP']=AaP
dff['Admin']=Admin
dff['Op_profit']=Op_profit
dff['EBITDA']=EBITDA
dff['EBITDA_margin']=EBITDA_margin
dff['Depreciation']=Depreciation
dff['Net_interest']=Net_interest
dff['Interest_recieved']=Interest_recieved
dff['Gross_debt']=Gross_debt
dff['Interest_paid']=Interest_paid
dff['EBT']=EBT
dff['EBT_after']=EBT_after
dff['Tax_rate']=Tax_rate
dff['PAT']=PAT
dff['Net_income']=Net_income
dff['Pay_out']=Pay_out
dff['Retained']=Retained
dff['Adjusted']=Adjusted
dff['EPS']=EPS
dff['DPS']=DPS
dff['Shareholder']=Shareholder
dff['BVPS']=BVPS
dff['CFPS']=CFPS

dff.round(1).T
```

```
Gross_margin=Gross_margin
Op_margin= Op_profit/df['Sales']*100
EBITDA_margin=EBITDA_margin
Pretax_margin=EBT/df['Sales']*100
Net_margin=PAT/df['Sales']*100
Opex_to_sales_ratio=-(df[' Advert,Promo & distibution '] + df['Admin cost'])/df['Sales']*100
ROAC=Net_income/Shareholder*100
Total_current_assets= df['Cash'] +df['Trade debtors'] + df['Other debtors']+df['Stock']+df['Prepayments and other current assets']
Total_non_current_assets= df['PP&E'] +df['Other non-current assets'] 
Total_assets=Total_current_assets+Total_non_current_assets
RoaA=Net_income/Total_assets*100
Debt_Equity=Gross_debt/Shareholder
Debt_EBITDA=Gross_debt/EBITDA
Invested_capital=Shareholder+Gross_debt
ROIC=Op_profit/Invested_capital*100
Total_current_liabilities=df['Trade creditors'] +df['Other creditors'] + df['Income tax payable']+df['ST debt']+df['Other Current Liabilities']
Current_ratio=Total_current_assets/Total_current_liabilities
Acid=(Total_current_assets-df['Stock'])/Total_current_liabilities*100
Net_debt=Gross_debt-df['Cash'] 
Net_debt_equity=Net_debt/Shareholder*100
Net_debt_EBITDA=Net_debt/EBITDA*100
Interest_cover=Op_profit/df['Interest paid']
Gross_debt_to_capital=Net_debt/EBITDA*100
No_days_receivables=df['Trade debtors']/df['Sales']*365
No_days_receivables_related=df['Other debtors']/df['Sales']*365
No_days_payables=df['Trade creditors']/df['Cost of sales']*365
No_days_payables_related=df['Other creditors']/df['Cost of sales']*365
Total_long_liabilities= df['LT debt'] + df['Deferred tax liability']+df['Other non-current liabilities']
Total_liabilities=Total_long_liabilities+Total_current_liabilities
Total_liability_equity=Total_liabilities+Shareholder+df['Minority Interest']
Net_debt_minority=Net_debt+df['Minority Interest1']
EBIT=Op_profit
D_A=-df['Depreciation/ Amortisation']  
working_capital=df['Net cash from operations']-EBIT-D_A-df['Income taxes paid']-df['Other operating activities']
Capex_turn_over=-df['Net Capex']/df['Sales']*100
Net_cash_investing=df['Net Capex']+df['Interest received']+df['Other investing activities']
Net_cash_to_firm=Net_cash_investing+df['Net cash from operations']
Net_cash_financing=df['Net debt proceeds']+df['Interest expense']+df['Share issue']+df['Other financing activity']+df['Dividend (Ordinary & Prefs)']
Net_increase_in_cash=Net_cash_financing+Net_cash_to_firm      
```

```
Ratios=pd.DataFrame(Gross_margin, columns = ['Gross_margin'])
Ratios['Op_margin']=Op_margin
Ratios['Pretax_margin']=Pretax_margin
Ratios['EBITDA_margin']=EBITDA_margin
Ratios['Net_margin']=Net_margin
Ratios['Opex_to_sales_ratio']=Opex_to_sales_ratio
Ratios['ROAC']=ROAC
Ratios['Total_current_assets']=Total_current_assets
Ratios['Total_non_current_assets']=Total_non_current_assets
Ratios['Total_assets']=Total_assets
Ratios['RoaA']=RoaA
Ratios['Debt_Equity']=Debt_Equity
Ratios['Debt_EBITDA']=Debt_EBITDA
Ratios['Invested_capital']=Invested_capital
Ratios['ROIC']=ROIC
Ratios['Total_current_liabilities']=Total_current_liabilities
Ratios['Current_ratio']=Current_ratio
Ratios['Acid']=Acid
Ratios['Net_debt']=Net_debt
Ratios['Interest_cover']=Interest_cover
Ratios['Gross_debt_to_capital']=Gross_debt_to_capital
Ratios['No_days_receivables']=No_days_receivables
Ratios['No_days_receivables_related']=No_days_receivables_related
Ratios['No_days_payables']=No_days_payables
Ratios['No_days_payables_related']=No_days_payables_related
Ratios['Total_long_liabilities']=Total_long_liabilities
Ratios['Total_liabilities']=Total_liabilities
Ratios['Total_liability_equity']=Total_liability_equity
Ratios['Net_debt_minority']=Net_debt_minority
Ratios['EBIT']=EBIT
Ratios['D_A']=D_A
Ratios['working_capital']=working_capital
Ratios['Capex_turn_over']=Capex_turn_over
Ratios['Net_cash_investing']=Net_cash_investing
Ratios['Net_cash_to_firm']=Net_cash_to_firm
Ratios['Net_cash_financing']=Net_cash_financing
Ratios['Net_increase_in_cash']=Net_increase_in_cash
Ratios.round(1).T
```
