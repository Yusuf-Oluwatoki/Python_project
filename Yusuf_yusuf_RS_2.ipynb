#Importing packages

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## IMPORTING DATA FROM LOCAL DRVE TO COLAB


#Creating a plot of each dataset

```
soil = pd.read_csv('Group12_soil.txt',delimiter = "\t")
soil=soil.set_index('USB2G28666_1:101')
soil=soil.drop('Unnamed: 2',axis=1)
soil=soil.rename(columns={'Unnamed: 1':'soil_Readings'})
soil.plot()
plt.grid()
plt.title("Plot of the Digital count and wavelength for soil_readings")
plt.ylabel('Digital count')
plt.xlabel('wavelength (nm)')
plt.grid()
plt.savefig('soil')
plt.show()
```

```
dark = pd.read_csv('Group12_dark.txt',delimiter = "\t")
dark=dark.set_index('USB2G28666_1:101')
dark=dark.drop('Unnamed: 2',axis=1)
dark=dark.rename(columns={'Unnamed: 1':'dark_Readings'})
dark.plot()
plt.grid()
plt.title("Plot of the Digital count and wavelength for dark")
plt.ylabel('Digital count')
plt.xlabel('wavelength (nm)')
plt.grid()
plt.savefig('dark')
plt.show()
```

```
both = pd.read_csv('Group12_both.txt',delimiter = "\t")
both=both.set_index('USB2G28666_1:101')
both=both.drop('Unnamed: 2',axis=1)
both=both.rename(columns={'Unnamed: 1':'both_Readings'})
both.plot()
plt.grid()
plt.title("Plot of the Digital count and wavelength for both")
plt.ylabel('Digital count')
plt.xlabel('wavelength (nm)')
plt.grid()
plt.savefig('both')
plt.show()
```

```
white = pd.read_csv('Group12_white.txt',delimiter = "\t")
white=white.set_index('USB2G28666_1:101')
white=white.drop('Unnamed: 2',axis=1)
white=white.rename(columns={'Unnamed: 1':'white_Readings'})
white.plot()
plt.grid()
plt.title("Plot of the Digital count and wavelength for white")
plt.ylabel('Digital count')
plt.xlabel('wavelength (nm)')
plt.grid()
plt.savefig('white')
plt.show()
```

```
grass = pd.read_csv('Group12_grass.txt',delimiter = "\t")
grass=grass.set_index('USB2G28666_1:101')
grass=grass.drop('Unnamed: 2',axis=1)
grass=grass.rename(columns={'Unnamed: 1':'grass_Readings'})
grass.plot()
plt.grid()
plt.title("Plot of the Digital count and wavelength for grass")
plt.ylabel('Digital count')
plt.xlabel('wavelength (nm)')
plt.grid()
plt.savefig('grass')
plt.show()
```

# Step 1: Combining all the spectrum collected.


```
All_spectrum=pd.concat([soil,both,dark,grass,white],axis=1)

All_spectrum
```

#  Step 2: Subtract the dark spectrum from all the other spectrum collected.


```
All_spectrum['soil_Readings']=All_spectrum.soil_Readings-All_spectrum.dark_Readings
All_spectrum['both_Readings']=All_spectrum.both_Readings-All_spectrum.dark_Readings
All_spectrum['white_Readings']=All_spectrum.white_Readings-All_spectrum.dark_Readings
All_spectrum['grass_Readings']=All_spectrum.grass_Readings-All_spectrum.dark_Readings

All_spectrum
```

#  Step 3: The downwelling irradiance derived from the radiance of the white reference by multiplying by pi: E d = L white *pi.


```
All_spectrum["downwelling"]=(All_spectrum.white_Readings*np.pi)
All_spectrum
```

#Step 3: Dividing all of the dark-subtracted radiance target spectra (grass, soil, and 50-50 mix) by the downwelling irradiance, e.g., R rs target = L target /(L white *pi).

```
All_spectrum["Soil_refl"]=All_spectrum.soil_Readings/All_spectrum.downwelling
All_spectrum["Grass_refl"]=All_spectrum.grass_Readings/All_spectrum.downwelling
All_spectrum["Both_refl"]=All_spectrum.both_Readings/All_spectrum.downwelling
All_spectrum
```

#Selecting wavelength between 400nm and 900nm


```
Plot_spectrum=All_spectrum[50:551]
Plot_spectrum
```

# Step 4: Plot all of the remote sensing reflectance target spectra on one plot..


```
plt.plot(All_spectrum.index,All_spectrum.Soil_refl,label='soil')
plt.plot(All_spectrum.index,All_spectrum.Grass_refl,label='grass')
plt.plot(All_spectrum.index,All_spectrum.Both_refl, label='both')
plt.ylim((0,1))
plt.legend(loc='upper center')
plt.grid()
plt.title("Plot of the Reflectance and wavelength for all spectra")
plt.ylabel('Reflectance ($sr^{-1}$)')
plt.xlabel('wavelength (nm)')
plt.grid()
plt.savefig('all')
plt.show()
```

```
plt.plot(Plot_spectrum.index,Plot_spectrum.Soil_refl,label='soil')
plt.plot(Plot_spectrum.index,Plot_spectrum.Grass_refl,label='grass')
plt.plot(Plot_spectrum.index,Plot_spectrum.Both_refl, label='both')
#plt.ylim((0,1))
plt.legend(loc='upper right')
plt.grid()
plt.title("Plot of the Reflectance and wavelength for all spectra")
plt.ylabel('Reflectance ($sr^{-1}$)')
plt.xlabel('wavelength (nm)')
plt.grid()
plt.savefig('all_400')
plt.show()
```

# Compared reflectance for derived and measured 

```
Plot_spectrum["Ratio_spectra"]=0.22*Plot_spectrum.Soil_refl+0.78*Plot_spectrum.Grass_refl
plt.plot(Plot_spectrum.index,Plot_spectrum.Ratio_spectra, label='Ratio')
plt.plot(Plot_spectrum.index,Plot_spectrum.Both_refl, label='both')
#plt.ylim((0,1))
plt.legend(loc='upper center')
plt.grid()
plt.title("Plot of the derived and measured Reflectance and wavelength for soil and grass")
plt.ylabel('Reflectance ($sr^{-1}$)')
plt.xlabel('wavelength (nm)')
plt.grid()
plt.savefig('compared.png')
plt.show()
```

#Error Analysis

```
Plot_spectrum['ERgrass']=0.05*Plot_spectrum.grass_Readings
Plot_spectrum['ERsoil']=0.05*Plot_spectrum.soil_Readings
Plot_spectrum['ERboth']=0.05*Plot_spectrum.both_Readings
Plot_spectrum['ERwhite']=0.15*Plot_spectrum.white_Readings
Plot_spectrum
```

```
Plot_spectrum['ergrass']=np.sqrt((Plot_spectrum.ERgrass/Plot_spectrum.downwelling)**2+((Plot_spectrum.grass_Readings*Plot_spectrum.ERwhite*np.pi)/(Plot_spectrum.downwelling**2))**2)
Plot_spectrum['ersoil']=np.sqrt((Plot_spectrum.ERsoil/Plot_spectrum.downwelling)**2+((Plot_spectrum.soil_Readings*Plot_spectrum.ERwhite*np.pi)/(Plot_spectrum.downwelling**2))**2)
Plot_spectrum['erboth']=np.sqrt((Plot_spectrum.ERboth/Plot_spectrum.downwelling)**2+((Plot_spectrum.both_Readings*Plot_spectrum.ERwhite*np.pi)/(Plot_spectrum.downwelling**2))**2)

Plot_spectrum
```

```
#Plot_spectrum["Ratio_spectra"]=0.22*Plot_spectrum.Soil_refl+0.78*Plot_spectrum.Grass_refl
plt.plot(Plot_spectrum.index,Plot_spectrum.Grass_refl, label='grass')
plt.plot(Plot_spectrum.index,(Plot_spectrum.Grass_refl-Plot_spectrum.ergrass), label='low')
plt.plot(Plot_spectrum.index,(Plot_spectrum.Grass_refl+Plot_spectrum.ergrass), label='high')
#ylim((0,1))
plt.legend(loc='upper center')
plt.grid()
plt.title("Plot of grass reflectance and its errors against wavelength")
plt.ylabel('Reflectance ($sr^{-1}$)')
plt.xlabel('wavelength (nm)')
plt.grid()
plt.savefig('ergrass.png')
plt.show()
```

```
#Plot_spectrum["Ratio_spectra"]=0.22*Plot_spectrum.Soil_refl+0.78*Plot_spectrum.Grass_refl
plt.plot(Plot_spectrum.index,Plot_spectrum.Soil_refl, label='soil')
plt.plot(Plot_spectrum.index,(Plot_spectrum.Soil_refl-Plot_spectrum.ersoil), label='low')
plt.plot(Plot_spectrum.index,(Plot_spectrum.Soil_refl+Plot_spectrum.ersoil), label='high')
#ylim((0,1))
plt.legend(loc='upper right')
plt.grid()
plt.title("Plot of soil reflectance and its errors against wavelength")
plt.ylabel('Reflectance ($sr^{-1}$)')
plt.xlabel('wavelength (nm)')
plt.grid()
plt.savefig('ersoil.png')
plt.show()
```

```
#Plot_spectrum["Ratio_spectra"]=0.22*Plot_spectrum.Soil_refl+0.78*Plot_spectrum.Grass_refl
plt.plot(Plot_spectrum.index,Plot_spectrum.Both_refl, label='both')
plt.plot(Plot_spectrum.index,(Plot_spectrum.Both_refl-Plot_spectrum.erboth), label='low')
plt.plot(Plot_spectrum.index,(Plot_spectrum.Both_refl+Plot_spectrum.erboth), label='high')
#ylim((0,1))
plt.legend(loc='upper center')
plt.grid()
plt.title("Plot of soil and grass reflectance and its errors against wavelength")
plt.ylabel('Reflectance ($sr^{-1}$)')
plt.xlabel('wavelength (nm)')
plt.grid()
plt.savefig('erboth.png')
plt.show()
```
