```python
# define the initial system state (aka x, y, z positions in space)
#initial_state = [1, 1, 1]

# define the system parameters sigma, rho, and beta

initial_state = [1200, 300, 0]
mu = 0.00025
gamma  = 0.25
beta  =  0.0013 
rho = (gamma / beta)


# define the time points to solve for, evenly spaced between the start and end times
start_time = 0
end_time = 100
time_points = np.linspace(start_time, end_time, end_time*100)
```

```python
# define the lorenz system
# x, y, and z make up the system state, t is time, and sigma, rho, beta are the system parameters
def diff_eqn_system(current_state, t):
    
    # positions of x, y, z in space at the current time point
    x_1, x_2, x_3 = current_state
    
    # define the 3 ordinary differential equations known as the lorenz equations
    dx1_dt = (-beta * x_1 * x_2) - (mu)
    dx2_dt = beta * x_2 * (x_1 - rho)
    dx3_dt = gamma * x_2
    
    # return a list of the equations that describe the system
    return [dx1_dt, dx2_dt, dx3_dt]
```

```python

# use odeint() to solve a system of ordinary differential equations
# the arguments are: 
# 1, a function - computes the derivatives
# 2, a vector of initial system conditions (aka x, y, z positions in space)
# 3, a sequence of time points to solve for
# returns an array of x, y, and z value arrays for each time point, with the initial values in the first row
x1_x_2_x_3 = odeint(diff_eqn_system, initial_state, time_points)

# extract the individual arrays of x, y, and z values from the array of arrays
x1 = x1_x_2_x_3[:, 0]
x2 = x1_x_2_x_3[:, 1]
x3 = x1_x_2_x_3[:, 2]
```

```python
figsize(10,5)

fig = plt.figure()
fig.patch.set_facecolor('xkcd:green')

plot(time_points,x1,color='red',label='x_1')
plot(time_points,x2,color='blue',label='x_2')
plot(time_points,x3,color='green',label='x_3')
xlabel('time')
ylabel('x_1/x_2/x_3 values')
legend(loc='upper right')
title('Plot of the Numerical solution of the given system of equation')
grid()



```

```python
# define the initial system state (aka x, y, z positions in space)
initial_state = [10, -10, 15]

# define the system parameters sigma, rho, and beta
sigma = 10.
rho   = 28.
beta  = 8./3.

# define the time points to solve for, evenly spaced between the start and end times
start_time = 0
end_time = 100
time_points = np.linspace(start_time, end_time, end_time*100)
```

```python
# define the lorenz system
# x, y, and z make up the system state, t is time, and sigma, rho, beta are the system parameters
def lorenz_system(current_state, t):
    
    # positions of x, y, z in space at the current time point
    x, y, z = current_state
    
    # define the 3 ordinary differential equations known as the lorenz equations
    dx_dt = sigma * (y - x)
    dy_dt = x * (rho - z) - y + x
    dz_dt = x * y - beta * z
    
    # return a list of the equations that describe the system
    return [dx_dt, dy_dt, dz_dt]
```

```python

# use odeint() to solve a system of ordinary differential equations
# the arguments are: 
# 1, a function - computes the derivatives
# 2, a vector of initial system conditions (aka x, y, z positions in space)
# 3, a sequence of time points to solve for
# returns an array of x, y, and z value arrays for each time point, with the initial values in the first row
xyz = odeint(lorenz_system, initial_state, time_points)

# extract the individual arrays of x, y, and z values from the array of arrays
x = xyz[:, 0]
y = xyz[:, 1]
z = xyz[:, 2]

```

```python
# define the initial system state (x, y positions in space)
initial_state = [10, 300]

# initialize the start time, end time and interval
start_time = 0
end_time = 1000
time_points = np.linspace(start_time, end_time, end_time*100)


# define the system of equation
def my_system(current_state, t):
    
    # assign x and y to the current position in time.
    x, y = current_state
    
    # define the 3 ordinary differential equations 
    dx_dt = 0.15 * (1 - (0.005*x) - (0.01*y)) * x
    dy_dt = 0.03 * (1 - (0.006*x) - (0.005*y)) * y
    
    # return a list of the equations that describe the system
    return [dx_dt, dy_dt]



# use odeint() to solve a system of ordinary differential equations
x1_x_2 = odeint(my_system, initial_state, time_points)

# extract the individual arrays of x, y, and z values from the array of arrays
xx1 = x1_x_2[:, 0]
xx2 = x1_x_2[:, 1]


#plot x and y against time
figsize(10,5)

fig = plt.figure()
fig.patch.set_facecolor('xkcd:green')
plot( time_points,xx1, color='red', alpha=1, linewidth=1,label='x')
plot(time_points,xx2, color='blue', alpha=1, linewidth=1,label='y')
xlabel('time')
ylabel('x/y values')
grid()
legend(loc='upper right')
title('Plot with starting values (10,10)')
```
