```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import kurtosis, skew, pearsonr, spearmanr
import seaborn as se
%matplotlib inline
```

# Question 1

```python
#Loading stock datas
Stock_1=pd.read_csv('Stock_1.csv')
Stock_2=pd.read_csv('Stock_2.csv')
```

```python
#Showing stock_1 data with its log_return
Stock_1.head(3)
```

```python
#Showing stock_2 data with its log_return
Stock_2.head(3)
```

```python
#Computing Statistics
Mean_stock_1=np.mean(Stock_1.log_return)
print("Mean_stock_1:%f" %Mean_stock_1)

Mean_stock_2=np.mean(Stock_2.log_return)
print("Mean_stock_2:%f" %Mean_stock_2)


Stand_Dev_stock_1=np.std(Stock_1.log_return)
print("Stand_Dev_stock_1:%f" %Stand_Dev_stock_1)

Stand_Dev_stock_2=np.std(Stock_2.log_return)
print("Stand_Dev_stock_2:%f" %Stand_Dev_stock_2)



Skewness_stock_1=skew(Stock_1.log_return)
print("Skewness_stock_1:%f" %Skewness_stock_1)

Skewness_stock_2=skew(Stock_2.log_return)
print("Skewness_stock_2:%f" %Skewness_stock_2)



Kurtosis_stock_1=kurtosis(Stock_1.log_return)
print("Kurtosis_stock_1:%f" %Kurtosis_stock_1)

Kurtosis_stock_2=kurtosis(Stock_2.log_return)
print("Kurtosis_stock_2:%f" %Kurtosis_stock_2)
```

```python
#Calculating covariance
covar=np.cov(Stock_1.log_return,Stock_2.log_return)
print(covar)
```

```python
#Calculating correlation
correl=pearsonr(Stock_1.log_return,Stock_2.log_return)
print("Pearsons correlation:%f" %correl[0])


correla=spearmanr(Stock_1.log_return,Stock_2.log_return)
print("Spaermans correlation:%f" %correla[0])
```

#Correlation Vs Covariance

Correlation and Covariance are very significant terms in the field of Staticstics. These two terms both measure the relationship and the dependency between two variables. The question is "What set them apart"?

Covariance indicates the direction of the linear relationship between variabbles while Correlation measure the strength and direction of the linear relationship. Hence Correlation values are standardized while that of Covariance are not.

```python
Stock_1=Stock_1.drop(['Year'], axis=1)
Stock_2=Stock_2.drop(['Year'], axis=1)
```

```python
#merging the two stock data and dividing into two uneven parts

Stock_1.rename(columns={'Adj Close':'Adj Close_1',
                          'log_return':'log_return_1'}, 
                 inplace=True)
Stock_2.rename(columns={'Adj Close':'Adj Close_2',
                          'log_return':'log_return_2'}, 
                 inplace=True)

All=pd.merge(Stock_1,Stock_2, how='inner', on='Date')
All.head(3)
```

```python
from sklearn.model_selection import train_test_split

## Spliting data into  80% and 20% 
Big_All, Small_All = train_test_split( All, test_size=0.20, random_state=42)
```
